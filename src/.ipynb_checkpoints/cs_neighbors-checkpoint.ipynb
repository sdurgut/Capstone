```python
dfGeo = pd.read_csv("../data/COUNTYCODES/CountyData.csv")
# drop some columns that are not interesting for our case.
dfGeo.drop(['Sort [1]','County Seat(s) [3]','Land Area mi²','Water Area mi²','Total Area mi²'], axis=1, inplace = True)
# fix the problematic non utf-8 characters
dfGeo['Latitude'] = dfGeo['Latitude'].apply(lambda x: float(x.split("°")[0].encode("utf-8")))
dfGeo['Longitude'] = dfGeo['Longitude'].apply(lambda x: float(x.split("°")[0].replace('–',"-").encode("utf-8")))
dfGeo.rename(columns={'County [2]': 'County'}, inplace=True)
dfGeo['County'] = dfGeo['County'].apply( lambda x : x.split('[')[0].strip() )
# convert FPIS codes to strings and make them 5 digits so that 1001 becomes 01001
dfGeo["FIPS"] = dfGeo["FIPS"].astype(str)
dfGeo["FIPS"] = dfGeo["FIPS"].str.zfill(5)
```

```python


myDf_ResultswithKalman_2YearIntervals = pd.read_csv("../data/RESULTS/ResultswithKalman_2YearIntervals.csv")
myDf_ResultswithKalman_2YearIntervals["FIPS"] = myDf_ResultswithKalman_2YearIntervals["FIPS"].astype(str)
myDf_ResultswithKalman_2YearIntervals["FIPS"] = myDf_ResultswithKalman_2YearIntervals["FIPS"].str.zfill(5)

myDf_Results =  pd.merge(myDf_ResultswithKalman_2YearIntervals, dfGeo, on='FIPS', how='inner')
myDf_Results.drop(["Population (2010)","County","State"],axis=1,inplace=True)


myDf_Results["Latitude"] = myDf_Results["Latitude"].astype(float)
myDf_Results["Longitude"] = myDf_Results["Longitude"].astype(float)
myDf_Results["ActualFatalityRate"] = myDf_Results["ActualFatalityRate"].astype(float)
myDf_Results = myDf_Results[np.isfinite(myDf_Results['Latitude'])]
myDf_Results = myDf_Results[np.isfinite(myDf_Results['Longitude'])]
myDf_Results["PredictionFatalityRate"] = myDf_Results["PredictionFatalityRate"].apply(lambda x: 0 if x < 0 else x)
```

## Using KNeighborRegressor

Unfortunately kneighborregressor did not help to improve the prediction score, so this part can be ignored

```python
myDf_Results["Pop_Actual"] = myDf_Results["Pop_Actual"].astype(float)
myDf_Results["Land Area km²"] = myDf_Results["Land Area km²"].astype(float)
myDf_Results["PopulationDensity"] = myDf_Results["Pop_Actual"]/myDf_Results["Land Area km²"]
# myDf_Results["Water Area km²"] = myDf_Results["Water Area km²"].astype(float)
# myDf_Results["Total Area km²"] = myDf_Results["Total Area km²"].astype(float)

# y_raw_train =
myDf_Results_train = myDf_Results[myDf_Results["Pop_Actual"]>25000]
myDf_Results_test = myDf_Results[myDf_Results["Pop_Actual"]<=25000]
X_train = myDf_Results_train[["Latitude","Longitude","PopulationDensity"]].copy()
X_test = myDf_Results_test[["Latitude","Longitude","PopulationDensity"]].copy()
y_train = myDf_Results_train["ActualFatalityRate"].copy()
y_test = myDf_Results_test["ActualFatalityRate"].copy()
np.nan_to_num(X_train)
np.nan_to_num(X_test)
np.nan_to_num(y_train)
np.nan_to_num(y_test)
print (X_train.shape,X_test.shape,y_train.shape,y_test.shape)
# myDf_Results
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
res = knn.predict(X_test)
# knn.score(X_test, y_test)
# np.where(np.isnan(X_test))
# len(res)
```

```python
y_test.shape
```

```python
prediction = myDf_Results_test["PredictionFatalityRate"]
resSeries = pd.Series(res)
# print (resSeries[:10],prediction[:10])
i = 0
for index,val in prediction.iteritems():
#     print (index,val) 
    tmp1 = 0
    tmp2 = 0
    if prediction[index] != np.NaN and prediction[index] > 0:
        tmp1 = prediction[index]
    if resSeries[i] != np.NaN and resSeries[i] > 0:
        tmp2 = resSeries[i] 
    prediction[index] = 0.7*tmp1+0.3*tmp2
    i+=1


```

```python
test = prediction
test1 = myDf_Results_train["PredictionFatalityRate"]
test = test.append(test1)
updatedPredictionFatalityRate = test.sort_index()
```

```python
myDf_Results["updatedPredictionFatalityRate"] = updatedPredictionFatalityRate
```
